```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('brent-monthly.csv')

# Data Exploration
print("Dataset Overview:")
print(data.head())
print("\nData Information:")
print(data.info())
print("\nSummary Statistics:")
print(data.describe())

# Data Visualization
sns.set(style='whitegrid')  # Set the style for Seaborn plots

# Example Plots

# Bar Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Date', y='Price', data=data)
plt.title('Bar Plot')
plt.show()

# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(data['Price'], bins=10)
plt.title('Histogram')
plt.show()

# Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Date', y='Price', data=data)
plt.title('Scatter Plot')
plt.show()

# Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Date', y='Price', data=data)
plt.title('Box Plot')
plt.show()

# Heatmap
correlation_matrix = data.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Heatmap')
plt.show()

# Pairplot
sns.pairplot(data, hue='Price')
plt.title('Pairplot')
plt.savefig('plot_name.png')
plt.show()
```
